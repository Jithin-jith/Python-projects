```python
from moviepy.video.io.VideoFileClip import VideoFileClip

def trim_video(input_file, time_ranges):
    """
    Trims a video file based on multiple start and end times and saves each segment as a separate file.

    :param input_file: Path to the input video file (MKV format).
    :param time_ranges: List of tuples containing start and end times in 'hh:mm:ss' format.
    """
    try:
        # Convert time to seconds
        def time_to_seconds(hhmmss):
            h, m, s = map(int, hhmmss.split(':'))
            return h * 3600 + m * 60 + s

        # Load the video file
        with VideoFileClip(input_file) as video:
            for i, (start_time, end_time) in enumerate(time_ranges, start=1):
                start_sec = time_to_seconds(start_time)
                end_sec = time_to_seconds(end_time)

                # Trim the video
                trimmed_clip = video.subclipped(start_sec, end_sec)

                # Generate output filename
                output_file = f"trimmed_part_{i}.mp4"

                # Write the trimmed video to the output file
                trimmed_clip.write_videofile(output_file, codec="libx264", audio_codec="aac")

                print(f"Trimmed video segment {i} saved to: {output_file}")

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
if __name__ == "__main__":
    input_path = "wWw.CinemaVilla.Mobi-Vikram Vedha (2017) Tamil HDRip 700MB x264 1CD MP3.mkv"  # Replace with your input MKV file path
    time_ranges = [
        ("00:07:18", "00:08:00")  # Replace with desired start and end times
    ]

    trim_video(input_path, time_ranges)
```
